```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

#print(Path("C:", "/", "real_estate.csv"))
ds = pd.read_csv("/Users/gfern/OneDrive/Escritorio/4GeeksAcademy/realestate-datacleanup-exercise/assets/real_estate.csv", sep=';')
ds['pps']=ds['price']/ds['surface']
Ciudades=['Fuenlabrada','Legans','Getafe','Alcorcn']
CSur= ds[ds['level5'].isin(Ciudades)]
#print(ds.head())
# Apply the default theme
sns.set_theme()

# Create a visualization
sns.relplot(
    data=CSur,
    x="pps", y="price", hue="level5",size="price")

sns.jointplot(data=CSur, x="pps", y="price", hue="level5")

#sns.pairplot(data=CSur, hue="level5")

```

```python
f, ax = plt.subplots(figsize=(7, 6))
sns.set_theme(style="ticks", palette="pastel")
#ax.set_xscale("log")
sns.boxplot(
    CSur, x="level5", y="price",whis=[0, 1], width=0.2, palette="vlag"
)
sns.despine(offset=10, trim=True)

```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# Plot a bar plot on the first subplot
sns.barplot(x="level5", y="price", data=CSur, ax=axs[0, 0])
axs[0, 0].set_title("Precios Cinturon Sur Madrid")

# Plot a box plot on the second subplot
sns.boxplot(x="price", y="level5", data=CSur,orient="h", ax=axs[0, 1])
axs[0, 1].set_title("Precios Cinturon Sur Madrid (Box Plot)")

# Plot a violin plot on the third subplot
sns.violinplot(x="level5", y="price", data=CSur, ax=axs[1, 0])
axs[1, 0].set_title("Precios Cinturon Sur Madrid (Violin Plot)")

# Plot a swarm plot on the fourth subplot
sns.swarmplot(x="level5", y="price", data=CSur, ax=axs[1, 1])
axs[1, 1].set_title("Precios Cinturon Sur Madrid (Swarm Plot)")

# Show the plot
plt.show()

print(CSur.shape)
```
